```python
XCOL = "body_mass_g"
YCOL = "bill_length_mm"
STYLE_FILE = "../src/style_files/nature.mplstyle"

PALETTE = [cm.managua(i / 2) for i in range(3)]

ROW_HEIGHT = 1.1           # vertical size per family
ASPECT = 4                 # width/height per facet
BW_ADJUST = 0.75            # KDE bandwidth scaling (smaller -> wigglier; bigger -> smoother)
BACKGROUND_COLOR = "0.25"  # light gray for overall family KDE
BACKGROUND_ALPHA = 0.5
PHASE_ALPHA = 0.95         # fill opacity for per-phase KDEs
LINEWIDTH = 0.9            # outline width
SHOW_RUG = False           # set True to add small ticks for samples
ORDER_FAMILIES_BY = "median"  # "median", "count", or "alpha" (alphabetical)
```

```python
sns.set_style("white")
sns.set_context("talk")
g = sns.JointGrid(
    data=penguins, x=XCOL, y=YCOL,
    height=2.5, ratio=4, space=0)


sns.kdeplot(data=penguins,  bw_adjust=1.11, cut =1.5,
            x=XCOL, y=YCOL, hue="species",
            palette=PALETTE, fill=True, alpha=0.75,
            common_norm=False,  ax=g.ax_joint)

sns.histplot(data=penguins, x=XCOL, hue="species", bins=50,
            palette=PALETTE, fill=True, alpha=0.75,
            legend=False, stat = "density", element="step", linewidth=LINEWIDTH/1.2,
            common_norm=False, multiple="stack", ax=g.ax_marg_x)


sns.histplot(data=penguins, y=YCOL, hue="species", bins=25,
            palette=PALETTE, fill=True, alpha=PHASE_ALPHA,
            legend=False, stat = "density", element="step", linewidth=LINEWIDTH/1.2,
            common_norm=False, multiple="stack", ax=g.ax_marg_y)
```

```python
plt.style.use(STYLE_FILE)

g = sns.JointGrid(
    data=penguins, x=XCOL, y=YCOL,
    height=2.5, ratio=4, space=0)


sns.kdeplot(data=penguins,  bw_adjust=1.11, cut =1.5,
            x=XCOL, y=YCOL, hue="species",
            palette=PALETTE, fill=True, alpha=0.75,
            common_norm=False,  ax=g.ax_joint, legend=False)

sns.histplot(data=penguins, x=XCOL, hue="species", bins=50,
            palette=PALETTE, fill=True, alpha=0.75,
            legend=False, stat = "density", element="step", linewidth=LINEWIDTH/1.2,
            common_norm=False, multiple="stack", ax=g.ax_marg_x)


sns.histplot(data=penguins, y=YCOL, hue="species", bins=25,
            palette=PALETTE, fill=True, alpha=PHASE_ALPHA,
            legend=False, stat = "density", element="step", linewidth=LINEWIDTH/1.2,
            common_norm=False, multiple="stack", ax=g.ax_marg_y)

g.set_axis_labels(xlabel="Body mass (g)", ylabel="Bill length (mm)")
```
