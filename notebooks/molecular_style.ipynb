```python
from rdkit import Chem
from rdkit.Chem import AllChem
import py3Dmol
```

```python
thiophene = "c1ccsc1"
mol = Chem.MolFromSmiles(thiophene)
Chem.AddHs(mol)
AllChem.Compute2DCoords(mol)
mol
```

```python
view = py3Dmol.view(width=800, height=800)
view.addModel(Chem.MolToMolBlock(mol), 'mol')
view.setStyle({'stick': {}, "sphere": {"scale": 0.3}})
view.show()
```

```python
element_colors = {
    "H": "#f2f2f2",   # Light Grey
    "C": "#D3D3D3",   # Dark Gray 
    "N": "#2944cf",   # Strong Blue
    "O": "#d60d0d",   # Bright Red
    "F": "#37c4a9",   # Sea Teal
    "Cl": "#41bf41",  # Fresh Green
    "P": "#FF8000",   # Standard Orange 
    "S": "#b5b50d",   # Old gold
}
```

```python
# display 3D structure from py3Dmol
view = py3Dmol.view(width=400, height=400)
view.addModel(Chem.MolToMolBlock(mol), 'mol')
# Set color for each atom
for elem, color in element_colors.items():
    view.setStyle({'elem': elem}, {'stick': {'color': color}})
view.setBackgroundColor('white')
# Add labels for each atom
for atom in mol.GetAtoms():
    idx = atom.GetIdx()
    symbol = atom.GetSymbol()
    view.addLabel(symbol,
        {'position': {'x': mol.GetConformer().GetAtomPosition(idx).x,
                      'y': mol.GetConformer().GetAtomPosition(idx).y,
                      'z': mol.GetConformer().GetAtomPosition(idx).z},
        'backgroundColor': 'white',
        'fontColor': 'black',
        'fontSize': 18,
        'showBackground': False,
        'fontOpacity': 0.5})
view.setViewStyle({
    'style':'outline',
    'color':'white',
    'width':0.05,
})
view.zoomTo()
view.show()
view.png()
```
